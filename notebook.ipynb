## 1. Importing Libraries

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

## 2. Loading Data from "Numpy Compressed Data Archive"

```python
data = np.load("./netflix_data.npz", allow_pickle=True)

P = data["P"]
Q = data["Q"]
bu = data["bu"]
bm = data["bm"]
mu = data["mu"]
movie_titles = data["movie_titles"]
movie_years = data["movie_years"]
```

## 3. Defining the Important Functions

```python
# Top Movies
def top_movies(bm, movie_titles, movie_years, n):
    sorted_index = np.argsort(bm)
    last_n = sorted_index[-n:]
    lines = []
    for i in last_n:
        lines.append(f"\nMovie Name: {movie_titles[i]}\nMovie Year: {movie_years[i]}")

    return "\n".join(lines)

# Search Movie
def search_movie(query, movie_titles, movie_years):
    query = query.lower()
    for index, movie in enumerate(movie_titles):
        movie = movie.lower()
        if query in movie:
            year = movie_years[index]
            return (index, movie, year)
    return (-1, "No Movie Found")

# Find Similar Movie
def find_similar_movies(movie_index, Q, top_n):
    target_vector = Q[movie_index]
    cosine_sim = cosine_similarity([target_vector], Q).flatten()
    sorted_array = np.argsort(cosine_sim)[::-1]
    top_indices = sorted_array[1:top_n + 1]
    result_list = []
    for index in top_indices:
        result_list.append(movie_titles[index])
    return "\n".join(result_list)
```

Similarity Function which we are using is nothing but a normal cosine similarity match.
$$\text{Cosine Similarity} = \cos(\theta) = \frac{A \cdot B}{||A|| \ ||B||}$$

## 4. Outputs

```python
# Shows top 10 movies
ans = top_movies(bm=bm, movie_titles=movie_titles, movie_years=movie_years, n=10)
print(ans)
```

```python
# Search movie
ans = search_movie("", movie_titles=movie_titles, movie_years=movie_years)

print(ans)
```

```python
# Find similar movies
ans1 = find_similar_movies(ans[0], Q=Q, top_n=10)

print(ans1)
```
